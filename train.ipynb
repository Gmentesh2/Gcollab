```python
import pandas as pd
import os
import matplotlib.pyplot as plt

```

```python
def symbol_to_path(symbol,base_dir="sample_data"):
  return os.path.join(base_dir,f"{symbol}.csv")
```

```python
symbols = ["DIS","FB","WMT"]
def get_data(symbols,dates):
  df = pd.DataFrame(index=dates)
  if "AAPL" not in symbols:
    symbols.insert(0,"AAPL")
  for symbol in symbols:
     main_df = pd.read_csv(
      symbol_to_path(symbol),
      index_col="Date",
      parse_dates=True,
      usecols=["Date","Adj Close"],
      na_values="nan"
      
  )   
     main_df = main_df.rename(columns={"Adj Close": symbol})
     df = df.join(main_df,how="inner")
     df = df.dropna()
     
  return df

 
dates = pd.date_range("2021-11-12","2021-11-23")
get_data(symbols,dates)
```
